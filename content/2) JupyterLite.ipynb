# Jupyter Lite

https://jupyterlite.readthedocs.io/en/latest/index.html
***

JupterLite is a JupyterLab instance, which runs in any browser, using JupyterLab components and extensions.
It is still an unoffical project, meaning not all the normal features from JupyterLab and and the classic notebooks will be available.


There are many advantages and uses for JupyterLite. <br>
As mentioned JupyterLite is a browser based deployment, this is done through Pyodide, JavaScript and P5.js. <br><br>
It allows you to view Notebooks and other files, as well as the ability to edit, delete, save and download copies of files from the browser.<br>
It also provides support for saving settings for JupyterLab and Lite. It also uses one kernel for one session.<br><br>
It is also very easily deployed, it is served via well-cacheable, static HTTP(S), locally or on most static web hosts
JupyterLite is also embeddable within larger applications, it also does not require a dedicated application server.

You can try a demo of JupyterLite before installing it, here:
https://jupyterlite.github.io/demo
***

## Installing JuypterLite

To add JupyterLite to your own github, simply create an empty repo with your username.github.io as the name of this repo, be sure to turn on github actions to activate github pages.

Next go to https://github.com/jupyterlite/demo and click the "Use this template" button, this will prompt you to add, to an existing or new repo, create a new repo, call it whatever you like. 

Now when you open your <b><i>username.github.io/jupyerliterepo</i></b> NOT the repo, it should start a JupyterLab/Lite deployment, allowing you to run juypter inside that page, features like creating new files etc, will all be available to you.
